```python
import os
import pandas as pd

# Define the path to your .txt files
folder_path = "/content/drive/MyDrive/Finance/txt_file"  # Corrected folder path

# Create a DataFrame to store post data
data = []

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
            lines = file.readlines()
            title = lines[0].strip()  # Assuming title (post owner) is on the first line
            body = " ".join([line.strip() for line in lines[1:]])  # The rest is body
            data.append({'title': title, 'body': body, 'filename': filename})

# Convert to DataFrame
df = pd.DataFrame(data)
```

Sentiment Analysis

```python
from textblob import TextBlob

# Function to get sentiment polarity
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity  # Returns a value between -1 (negative) and 1 (positive)

# Apply sentiment analysis to the post bodies
df['sentiment'] = df['body'].apply(get_sentiment)

# Label posts as positive, neutral, or negative based on sentiment score
df['sentiment_label'] = pd.cut(df['sentiment'], bins=[-1, -0.1, 0.1, 1], labels=['negative', 'neutral', 'positive'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar chart for sentiment distribution
plt.figure(figsize=(8,6))
sns.countplot(x='sentiment_label', data=df, palette='coolwarm')
plt.title('Sentiment Distribution of Posts')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

Topic Modeling

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Convert posts to a document-term matrix
vectorizer = CountVectorizer(stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df['body'])

# Fit LDA model to discover 5 topics
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(doc_term_matrix)

# Get the top words for each topic
words = vectorizer.get_feature_names_out()
topics = []
for topic_idx, topic in enumerate(lda.components_):
    top_words = [words[i] for i in topic.argsort()[-10:]]
    topics.append(f"Topic {topic_idx + 1}: {', '.join(top_words)}")

df['topics'] = lda.transform(doc_term_matrix).argmax(axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define topic labels specific to financial analysis
topic_labels = {
    0: 'Market Trends',
    1: 'Stock Performance',
    2: 'Currency Exchange Rates',
    3: 'Government Regulations',
    4: 'Investor Sentiments'
}

# Map the numeric topics to their corresponding labels
df['topic_names'] = df['topics'].map(topic_labels)

# Bar chart for topic distribution with custom labels
plt.figure(figsize=(10,7))  # Set figure size
sns.countplot(x='topic_names', data=df, palette='viridis', edgecolor='black')  # Use mapped topic names

# Add titles and labels with improved font sizes
plt.title('Topic Distribution of Financial Reports', fontsize=16, fontweight='bold')
plt.xlabel('Topics', fontsize=14)
plt.ylabel('Number of Reports', fontsize=14)

# Rotate x-axis labels if needed for better readability
plt.xticks(rotation=45, fontsize=12)

# Add gridlines for the y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot with a tight layout
plt.tight_layout()
plt.show()
```

```python
from spellchecker import SpellChecker

# Initialize the spell checker
spell = SpellChecker()

# Function to count spelling mistakes
def count_spelling_mistakes(text):
    words = text.split()
    misspelled = spell.unknown(words)
    return len(misspelled)

# Function to detect carelessly written posts
def is_careless(text):
    mistakes = count_spelling_mistakes(text)
    return mistakes > 5  # Arbitrary threshold for spelling mistakes

# Apply the function
df['carelessly_written'] = df['body'].apply(is_careless)
```

# **Detecting Personal Opinions**

```python
# Function to detect personal opinions based on first-person pronouns
def is_personal_opinion(text):
    personal_pronouns = ['I', 'me', 'my', 'we', 'us', 'our']
    return any(pronoun in text.split() for pronoun in personal_pronouns)

# Apply the function
df['personal_opinion'] = df['body'].apply(is_personal_opinion)
```

# **Detecting New Posts**

```python
# Function to detect news-like posts based on formal tone
def is_news_like(text):
    news_pronouns = ['he', 'she', 'they', 'it', 'the']
    return any(pronoun in text.split() for pronoun in news_pronouns)

# Apply the function
df['news_like'] = df['body'].apply(is_news_like)
```

# **Formal and Informal Detection**

```python
import os
import pandas as pd

# Define the path to your .txt files
folder_path = "/content/drive/MyDrive/Finance/txt_file"
# Read the content of each .txt file into a DataFrame
data = []
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
            post_title = file.readline().strip()  # Assuming the first line is the title
            post_body = file.read().strip()  # The rest is the body
            data.append({'title': post_title, 'body': post_body})

df = pd.DataFrame(data)
```

```python
import spacy
import re

# Load spaCy's English model
nlp = spacy.load('en_core_web_sm')

# Function to detect informal language based on certain features
def detect_informal_language(text):
    doc = nlp(text)
    informal_features = 0

    # Feature 1: Use of contractions (common in informal language)
    contractions = re.findall(r"\b(can't|won't|ain't|I'm|you're|we're|they're)\b", text.lower())
    informal_features += len(contractions)

    # Feature 2: Use of slang words
    slang_words = ['gonna', 'wanna', 'kinda', 'sorta', 'y\'all', 'dunno']
    informal_features += sum([1 for word in slang_words if word in text.lower()])

    # Feature 3: Exclamation marks (can indicate informal tone)
    informal_features += text.count('!')

    # Feature 4: Overuse of pronouns (common in informal speech)
    pronouns = [token for token in doc if token.pos_ == 'PRON']
    informal_features += len(pronouns)

    # Classify as informal if there are more than 2 informal features
    if informal_features > 2:
        return 'Informal'
    else:
        return 'Formal'

# Apply the function to each post body
df['tone'] = df['body'].apply(detect_informal_language)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of formal vs. informal posts
plt.figure(figsize=(8, 6))
sns.countplot(x='tone', data=df, palette='coolwarm')
plt.title('Distribution of Formal vs. Informal Posts')
plt.xlabel('Tone')
plt.ylabel('Count')
plt.show()
```

```python
# Define a function to assign topics based on keywords in a 'body' column
def assign_topic(text):
    if "economy" in text.lower():
        return "Economy"
    elif "politics" in text.lower():
        return "Politics"
    elif "reform" in text.lower():
        return "Reform"
    elif "devaluation" in text.lower():
        return "Devaluation"
    elif "foreign exchange" in text.lower():
        return "Foreign Exchange"
    else:
        return "Other"

# Apply the function to create a 'topics' column
df['topics'] = df['body'].apply(assign_topic)

# Now, group by 'tone' and 'topics'
tone_topic = df.groupby(['tone', 'topics']).size().unstack()

# Plot a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(tone_topic, annot=True, cmap='Blues')
plt.title('Tone vs. Topic Correlation')
plt.xlabel('Topics')
plt.ylabel('Tone')
plt.show()
```

# **Model Training**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import re # Added missing import

# Sample preprocessing function
def preprocess_text(text):
    # Example preprocessing (lowercasing, removing punctuation, etc.)
    text = text.lower()
    text = re.sub(r'\W+', ' ', text)
    return text

# Apply preprocessing
df['processed_body'] = df['body'].apply(preprocess_text)

# Assume 'tone' column exists with the category labels - 'labels' was not found
X = df['processed_body']
y = df['tone']  # Replace 'tone' with your actual column name for labels if it's not 'tone'

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

Train **Model**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

# Create a pipeline with TF-IDF and a classifier
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])

# Train the model
pipeline.fit(X_train, y_train)

# Predict on the test set
y_pred = pipeline.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
# Example of classifying new data
new_posts = ["Sample post text to classify", "Another example post"]

# Preprocess and classify
processed_new_posts = [preprocess_text(post) for post in new_posts]
predicted_labels = pipeline.predict(processed_new_posts)

# Decode the labels
predicted_labels_decoded = le.inverse_transform(predicted_labels)

print(predicted_labels_decoded)
```
